```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
df.shape
```

```python
df.dtypes
```

```python
df = df.drop(['id','spkid','full_name','pdes','prefix','name'],axis=1)
df = df.dropna()
df
```

```python
df.dtypes
```

```python
le1 = LabelEncoder() 
le2 = LabelEncoder() 
le3 = LabelEncoder() 
le4 = LabelEncoder() 
le5 = LabelEncoder() 

le1.fit(df['neo'])
le2.fit(df['pha'])
le3.fit(df['orbit_id'])
le4.fit(df['equinox'])
le5.fit(df['class'])

df['neo'] = le1.transform(df['neo'])
df['pha'] = le2.transform(df['pha'])
df['orbit_id'] = le3.transform(df['orbit_id'])
df['equinox'] = le4.transform(df['equinox'])
df['class'] = le5.transform(df['class'])

df.head()
```

```python
X = df.drop(['pha'], axis=1)
y = df['pha']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
model = LogisticRegression()

model.fit(X_train, y_train)

answers_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Как оценить насколько хорошо работает ваша модель?

# answers_pred - ответы которые вернула модель для X_test
# y_test - это правильные ответы для X_test
print(accuracy_score(y_test,answers_pred))
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

confusion_matrix(y_test, answers_pred)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
corpus = [
          'Доброта - это стремление помочь людям, причём не требуя за это благодарности',
          'Это свойство души, позволяющее не оставаться равнодушным к бедам других, оказываться рядом тогда, когда это так необходимо человеку',
          'Веди без доброты человек не сможет жить как человек',
          'Благодаря доброте мы радуемся и не только',
]

X = vectorizer.fit_transform(corpus)
X.toarray()
```

```python
import pandas as pd

df = pd.DataFrame(data =X.toarray(), columns=vectorizer.get_feature_names())
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = [
          'Доброта - это стремление помочь людям, причём не требуя за это благодарности',
          'Это свойство души, позволяющее не оставаться равнодушным к бедам других, оказываться рядом тогда, когда это так необходимо человеку',
          'Веди без доброты человек не сможет жить как человек',
          'Благодаря доброте мы радуемся и не только',
]
# Создаем модель векторизации текста
vectorizer = TfidfVectorizer()

# токенезация текста и построение словоря
vectorizer.fit(corpus)

# summarize
print(vectorizer.vocabulary_)
print(vectorizer.idf_)
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)

feature_names = vectorizer.get_feature_names()

dense = vectors.todense()

denselist = dense.tolist()

df2 = pd.DataFrame(denselist, columns=feature_names)
df2
```

```python
import pandas as pd
df = pd.DataFrame(["Мама мыла раму, потом пошла обедать.", "Рама мыла маму, что очень странно.", "Мыло мыло маму, забавно получается.", "Раму мыла мама, а я гулял."], columns = ['text'])
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

corpus = [
          'Text of first document Text Text.',
          'Text of the second document made longer.',
          'Number three.',
          'This is number four.',
]

X = vectorizer.fit_transform(df.text)

print(vectorizer.get_feature_names())

X.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

# токенезация текста и построение словоря
vectorizer.fit(df.text)

# summarize
print(vectorizer.vocabulary_)
print(vectorizer.idf_)
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(df.text)

feature_names = vectorizer.get_feature_names()
print("Feature names", feature_names)

dense = vectors.todense()

denselist = dense.tolist()

df2 = pd.DataFrame(denselist, columns=feature_names)
df2
```

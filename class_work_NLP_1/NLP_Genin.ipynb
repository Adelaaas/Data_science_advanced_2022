```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
vectorizer = CountVectorizer()
corpus=["Файл CSV может содержать несколько строкa после строкa заголовка.",
        "Каждая строка соответствует артефакт, который будет создан. Запись в столбце Тип соответствует тип создаваемого артефакт в системе.",
        "Остальные ячейки совпадают с соответствующими полями артефакт."]
X=vectorizer.fit_transform(corpus)
X.toarray()
df=pd.DataFrame(data =X.toarray(), columns=vectorizer.get_feature_names())
df['Document']=pd.Series(corpus)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
corpus=["Файл может содержать несколько строкa после строка заголовка.",
        "Каждая строкa соответствует артефакт, который будет создан. Запись в столбце Тип соответствует тип создаваемого артефакт в системе.",
        "Остальные ячейки совпадают с соответствующими полями артефакт строкa."]
vectorizer.fit(corpus)
print(vectorizer.vocabulary_)
print(vectorizer.idf_)
```

```python
vectors = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names()
df2 = pd.DataFrame(vectors.todense().tolist(), columns=feature_names)
df2['Document']=pd.Series(corpus)
df2
```

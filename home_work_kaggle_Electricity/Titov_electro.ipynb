```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("sample.csv")
```

```python
train.info()
```

```python
train.isna().sum()
```

```python
train = train.dropna().reset_index()
```

```python
train.info()
```

```python
train['h'] = 0
train['d'] = 0
train['m'] = 0
train['y'] = 0
```

```python
train
```

```python
for i in range(34946):
    train['h'][i] = train["datetime"][i][11:13]
    train['d'][i] = train["datetime"][i][0:2]
    train['m'][i] = train["datetime"][i][3:5]
    train['y'][i] = train["datetime"][i][6:10]
  
```

```python
train = train.drop(["datetime"], axis = 1)
train = train.drop(["index"], axis = 1)
```

```python
train
```

```python
from sklearn.linear_model import LinearRegression

x = train[['h', 'd', 'm', 'y']]
y = train[['total']]
```

```python
pres = LinearRegression()
pres.fit(x,y)

predict_test = pres.predict(x)
predict_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error


print('MSE: {:.3f}'.format(mean_squared_error(y, predict_test)))
```

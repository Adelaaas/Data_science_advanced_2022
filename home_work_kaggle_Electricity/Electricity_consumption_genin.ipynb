```python
from google.colab import files
uploaded=files.upload()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train=pd.read_csv("train.csv")
sample=pd.read_csv("sample.csv")
```

```python
test=sample.copy()
print("Тренировочный:",train.shape)
print("Тестовый:",test.shape)
print("_____________")
print("Тренировочный:")
print(train.info())
print(" ")
print("Тестовый:")
print(test.info())
print("_____________")
print("Тренировочный:")
print(train.isna().sum())
print(" ")
print("Тестовый:")
print(test.isna().sum())
```

```python
train.dropna(axis=0,how='any',inplace=True)
train.shape
```

```python
test
```

```python
train['День']=0
train['Месяц']=0
train['Час']=0
for index, row in train.iterrows():
  train.loc[index,'День']=int(train.loc[index,'datetime'][:2])
  train.loc[index,'Месяц']=int(train.loc[index,'datetime'][3:5])
  train.loc[index,'Час']=int(train.loc[index,'datetime'][11:13])
train.drop('datetime',axis=1,inplace=True)
train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
x = train.drop(columns=["total"])
y = train["total"]
x_tr, x_te, y_tr, y_te = train_test_split(x, y, random_state=42, test_size=0.3)
lire = LinearRegression()
lire.fit(x_tr,y_tr)
pred=lire.predict(x_te)
print('MSE: {:.3f}'.format(mean_squared_error(y_te, pred)))
```

```python
test['День']=0
test['Месяц']=0
test['Час']=0
for index, row in test.iterrows():
  test.loc[index,'День']=int(test.loc[index,'datetime'][:2])
  test.loc[index,'Месяц']=int(test.loc[index,'datetime'][3:5])
  test.loc[index,'Час']=int(test.loc[index,'datetime'][11:13])
test.drop('datetime',axis=1,inplace=True)
test
```

```python
test.drop(columns=['total'],inplace=True)
sampred=pd.Series(lire.predict(test))
sample['total']=sampred
sample
```

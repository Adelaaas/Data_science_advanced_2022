```python
import pandas as pd
df = pd.read_csv("/content/train.csv")
df
```

```python
df = df.drop(0)
df
```

```python
for i in range(1, 35064):
  a = df.loc[i, 'datetime']
  df.loc[i, 'day'] = a[0:2]
  df.loc[i, 'month'] = a[3:5]
  df.loc[i, 'year'] = a[6:10]
  df.loc[i, 'hours'] = a[11:13]
df = df.drop(columns = ['datetime'])
df 
```

```python
y = df['total']
df = df.drop(columns = ['total'])
```

```python
import numpy as np
fill_value = np.mean(y)
y = y.fillna(fill_value)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y)
y = le.transform(y)
y
```

```python
#MinMaxScaler не подходит
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_st = scaler.fit_transform(df)
pd.DataFrame(X_st)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, KFold
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_st, y)
knn_preds = knn.predict(X_st)
knn_preds_proba = knn.predict_proba(X_st)[:, 1]

print('accuracy', accuracy_score(y, knn_preds))
```

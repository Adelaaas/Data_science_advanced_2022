```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, plot_confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('predict.csv')
```

```python
import re 
#re.findall(r'[A-z]\d([A-Za-z]*)', 'A0V')
train['first_bukva'] = 0
train['first_classification'] = 0
train['second_classification'] = 0
for index, row in train.iterrows():
  train.loc[index, 'first_bukva'] = re.split(r'\d', train.loc[index, 'SpType'])[0]
  try:
    a = re.findall(r'.*[0-9]([IV]*)', train.loc[index, 'SpType'])[-1]
    if a  not in ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII']:
      train.loc[index, 'first_classification'] = 'IV'
    else:
      train.loc[index, 'first_classification'] = a


    #print(index)
    #print(re.findall(r'.*\d', train.loc[index, 'SpType']))
    train.loc[index, 'second_classification'] = re.findall(r'.([0-9])', train.loc[index, 'SpType'])[0]
  except IndexError:
      train.loc[index, 'first_classification'] = 'IV'
      train.loc[index, 'second_classification'] = 4

  #print(re.split(r'\d[A-Za-z]*', i)).*\d
```

```python
train = train.drop(['SpType'], axis=1)
train
```

```python
from sklearn import preprocessing
 

l1 = preprocessing.LabelEncoder()
l1.fit(train['first_bukva'])
train['first_bukva'] = l1.transform(train['first_bukva'])

l2 = preprocessing.LabelEncoder()
l2.fit(train['first_classification'])
train['first_classification'] = l2.transform(train['first_classification'])

'''l3 = preprocessing.LabelEncoder()
l3.fit(train['second_classification'])
train['second_classification'] = l3.transform(train['second_classification'])
'''

train
```

```python
#from sklearn.preprocessing import MinMaxScaler

X = train.drop(['TargetClass'], axis=1)
y = train['TargetClass']
'''
scaler = MinMaxScaler([0,1])
X_n = scaler.fit_transform(X)
X_n'''
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 1/3)
```

```python
rfc = RandomForestClassifier(random_state = 42, n_jobs = -1)
rfc.fit(X_train, y_train)
```

```python
pred = rfc.predict(X_test)
```

```python
roc_auc_score(y_test, pred)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, plot_confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import re
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_advanced_2022/main/%D0%9D%D0%A2%D0%9E%20%D0%B8%D0%BD%D0%B6%20%D1%82%D1%83%D1%80/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_advanced_2022/main/%D0%9D%D0%A2%D0%9E%20%D0%B8%D0%BD%D0%B6%20%D1%82%D1%83%D1%80/predict.csv')
```

```python
train['Class'] = 0
train['illum'] = 0


for index, row in train.iterrows():
    train.loc[index, 'Class'] = re.findall(r'([A-Z])', train.loc[index, 'SpType'])[0]
    try:
        x = re.findall(r'([0-9])', train.loc[index, 'SpType'])[0]
        y = re.findall(r'.*[0-9]([IV]*)', train.loc[index, 'SpType'])[-1]
        if y not in ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII']:
            train.loc[index, 'illum'] = 'IV'
        else:
            train.loc[index, 'illum'] = y
        train.loc[index, 'class_class'] = x
    except IndexError:
        train.loc[index, 'illum'] = 'IV'
        train.loc[index, 'class_class'] = 4
sptype = train.drop("SpType", axis=1, inplace=True)

train
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
label1 = le1.fit_transform(train['Class'])
label2 = le2.fit_transform(train['illum'])
train.drop("Class", axis=1, inplace=True)
train.drop("illum", axis=1, inplace=True)
train['Class'] = label1
train['illum'] = label2

train
```

```python
y = train['TargetClass']
X = train.drop(columns = ['TargetClass'])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.3)
```

```python
%%time
rfc = RandomForestClassifier(random_state = 42, n_jobs = -1)
rfc.fit(X_train, y_train)
```

```python
pred = rfc.predict(X_test)
roc_auc_score(y_test, pred)
```
